```python
# read noisy data from csv in pandas

import pandas as pd
import numpy as np

# read data from csv
c = pd.read_csv('iawe/noisy/c_final_noisy.csv')
tv = pd.read_csv('iawe/noisy/tv_final_noisy.csv')
uk = pd.read_csv('iawe/noisy/uk_final_noisy.csv')
```

```python
# train LSTM on 20% of noisy data

# split data into input and output
length = len(c)
p1 = int(0.2 * length)

c_input = c.iloc[:, 0:p1]
c_output = c.iloc[:, p1:length]

tv_input = tv.iloc[:, 0:p1]
tv_output = tv.iloc[:, p1:length]

uk_input = uk.iloc[:, 0:p1]
uk_output = uk.iloc[:, p1:length]   

# convert to numpy array
c_input = c_input.to_numpy()
tv_input = tv_input.to_numpy()
uk_input = uk_input.to_numpy()

c_output = c_output.to_numpy()
tv_output = tv_output.to_numpy()
uk_output = uk_output.to_numpy()

# reshape input to 3D array
c_input = c_input.reshape(c_input.shape[0], c_input.shape[1], 1)    
tv_input = tv_input.reshape(tv_input.shape[0], tv_input.shape[1], 1)
uk_input = uk_input.reshape(uk_input.shape[0], uk_input.shape[1], 1)

# reshape output to 3D array
c_output = c_output.reshape(c_output.shape[0], c_output.shape[1], 1)
tv_output = tv_output.reshape(tv_output.shape[0], tv_output.shape[1], 1)
uk_output = uk_output.reshape(uk_output.shape[0], uk_output.shape[1], 1)

# train LSTM on 20% of noisy data
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout

# create LSTM model
def create_model():
    model = Sequential()
    model.add(LSTM(units=50, return_sequences=True, input_shape=(c_input.shape[1], 1)))
    model.add(Dropout(0.2))
    model.add(LSTM(units=50, return_sequences=True))
    model.add(Dropout(0.2))
    model.add(LSTM(units=50, return_sequences=True))
    model.add(Dropout(0.2))
    model.add(LSTM(units=50))
    model.add(Dropout(0.2))
    model.add(Dense(units=1))
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

# fit model
model1 = create_model()
model1.fit(c_input, c_output, epochs=100, batch_size=32)

model2 = create_model()
model2.fit(tv_input, tv_output, epochs=100, batch_size=32)

model3 = create_model()
model3.fit(uk_input, uk_output, epochs=100, batch_size=32)

# save model
model1.save('iawe/noisy/c_model.h5')
model2.save('iawe/noisy/tv_model.h5')
model3.save('iawe/noisy/uk_model.h5')

# test LSTM forecast on 80% of noisy data

# split data into input and output
p2 = int(0.8 * length)

c_input = c.iloc[:, 0:p2]
c_output = c.iloc[:, p2:length]

tv_input = tv.iloc[:, 0:p2]
tv_output = tv.iloc[:, p2:length]


    




```
